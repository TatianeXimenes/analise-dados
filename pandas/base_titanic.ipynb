### 5. Consulta dos dados

#### 5.1 com o loc

Com o método loc podemos selecionar um subconjunto de linhas e colunas de um DataFrame usando a indexação baseada em um rótulo (título das colunas).

```python
#Retorne todas as informações do 1º passageiro
df.loc[1]
```

```python
#Retorne todas as informações dos passageiros 1, 2 e 3, utilizando uma lista.
df.loc[[1,2,3]]
```

```python
#Selecionando vários indices e colunas atráves de duas listas
df.loc[[1,2], ['Name','Sex','Age']]
```

```python
#Fazendo fatiamento (slice). 
#Podemos pedir para que seja pulado de 2 em 2: df.loc[start : stop : step].
#Podemos pedir apenas os 10 primeiros passageiros: df.loc[10:]
df.loc[10:20]
```

```python
#Slice de linhas e colunas
df.loc[1:10, ['Name','Sex','Age']]
```

#### 5.2 com o query

```python
#Retorna os 5 primeiros passageiros maiores de 20 anos
df.query('Age > 20').head()
```

```python
#Retorna os 10 primeiros passageiros maiores de 20 anos desde que sejam homens
df.query('Age > 20 & Sex=="male"').head(10)
```

```python
#Retorna as 10 primeiras pessoas que sejam maiores de 20 anos OU pessoas do sexo masculino
df.query('Age > 20 | Sex=="male"').head(10)
```

```python
#Retorna os 5 primeiros se o valor da coluna Embarked é igual a C ou Q
df.query('Embarked in ["C","Q"]', inplace=True)
df.head()
```

### 10. Pandas para Manipulação de Strings:

```python
#O método rstrip() remove os espaços à direita, enquanto o lstrip() remove espaços à esquerda da string. 
df.Name.str.rstrip().head()
```

```python
#Para transformar todo texto em minúsculo usa-se o método lower(), para maiúsculo o método upper().
df.Name.str.upper().head()
```

```python
#Função para remover todos os parênteses, o replace substitui os parênteses por vázio.
def remove_parenteses(item):
  if '(' in item:
    return item.replace('(','').replace(')','')
  else:
    return item

# O método apply(), permite que a função criada anteriormente seja aplicada em todas as linhas da coluna Name.
#Dica: Para aplicar essa mesma função para várias colunas você deve usar o applymap().
df.Name.head(10).apply(remove_parenteses)
```

```python
#Expressão regular é uma sequência de caracteres usados principalmente para encontrar e substituir padrões numa string ou num arquivo.
import re
```

```python
#5 primeiros passageiros que tem ‘Mr’ no nome
#O método flags=re.I é usado para ignorar tanto valores minúsculos quanto maiúsculos
df.loc[df.Name.str.contains('Mr', flags=re.I, regex=True)].head()
```

### 11. Dicas para trabalhar com grandes bases de dados com Pandas:

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```
